```python
df['BASIC_PAY'].fillna(df['BASIC_PAY'].mean(), inplace=True)
```

```python
df['BASIC_PAY'] = df['BASIC_PAY'].fillna(df['BASIC_PAY'].mean())
```

```python
df['HRA'] =df['HRA'].fillna(df['HRA'].mean())
```

```python
df['DA'] = df['DA'].fillna(df['DA'].mean())
```

```python
df['TAX'].fillna(df['TAX'].mean(),inplace=True)
```

```python
print("Dataset after handling missing values:")
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[['BASIC_PAY']] = scaler.fit_transform(df[['BASIC_PAY']])
```

```python
df['HRA'] = (df['HRA'] - df['HRA'].min()) / (df['HRA'].max() - df['HRA'].min())
```

```python
from sklearn.preprocessing import StandardScaler

s_scaler = StandardScaler()
```

```python
df[['DA']] = s_scaler.fit_transform(df[[ 'DA']])
```

```python
df['DA']
```

```python
df['DA'] = (df['DA'] - df['DA'].mean()) / df['DA'].std()
```
